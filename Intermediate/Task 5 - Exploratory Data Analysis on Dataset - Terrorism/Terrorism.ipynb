```python
# Importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("globalterrorismdb_0718dist.csv", encoding = "ISO-8859-1")
df
```

```python
df.columns.values
```

```python
df.describe(include="all")
```

```python
df.shape
```

```python
print("Maximum Terrorism attacked on Country         :", df["country_txt"].value_counts().idxmax())
print("City most attacked by Terrosist               :",df['city'].value_counts().index[1])
print("Highly Terrorism Region                       :", df["region_txt"].value_counts().index[0])
print("Year on which maximum terrorism occured       :", df["iyear"].value_counts().idxmax())
print("Month on which maximum attacks are done       :",df['imonth'].value_counts().idxmax())
print("Day on which maximum terrirsm attack are done :",df['iday'].value_counts().idxmax())
print("Group with the most attacks                   :",df['gname'].value_counts().index[1])
print("Most Attack Types                             :",df['attacktype1_txt'].value_counts().idxmax())
print("Most Target Types                             :",df['targtype1_txt'].value_counts().idxmax())
print("Successful Terrorist attack                   :",df["success"].value_counts().sum())
```

```python
plt.subplots(figsize=(10,6))
sns.countplot('country_txt',data=df,order= df['country_txt'].value_counts().index[:10])
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.title('Maximum Number Of Terrorist Activities By Country')
plt.show()
```

```python
plt.subplots(figsize=(10,6))
sns.countplot('city',data=df,order= df['city'].value_counts().index[1:10])
plt.xticks(rotation=90)
plt.xlabel("City")
plt.title(' Maximum Number Of Terrorist Activities By City')
plt.show()
```

```python
plt.subplots(figsize=(10,6))
sns.countplot('region_txt',data=df,order= df['region_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel("Regions")
plt.title('Maximum Number Of Terrorist Activities By Region')
plt.show()
```

```python
plt.subplots(figsize=(10,6))
sns.countplot('iyear',data=df,order= df['iyear'].value_counts().index[:10])
plt.xticks(rotation=90)
plt.xlabel("year")
plt.title('Maximum Number Of Terrorist Activities By year')
plt.show()
```

```python
plt.subplots(figsize=(16,6))
sns.countplot(df['imonth'], data=df,order= df['imonth'].value_counts().index[:12])
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.xlabel("Month")
plt.title('Terrorism Activities Each Month')
plt.show()
```

```python
plt.subplots(figsize=(16,6))
sns.countplot(df['iday'], data=df,order= df['iday'].value_counts().index[:5])
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.xlabel("Day")
plt.title('Maximum Terrorism Activities By Day')
plt.show()
```

```python
plt.subplots(figsize=(16,6))
sns.countplot(df['gname'], data=df,order= df['gname'].value_counts().index[1:5])
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.xlabel("Group")
plt.title('Maximum Terrorism Activities By Group')
plt.show()
```

```python
plt.subplots(figsize=(16,6))
sns.countplot(df['attacktype1_txt'], data=df,order= df['attacktype1_txt'].value_counts().index[:5])
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.xlabel("Attack Type")
plt.title('Maximum Terrorism Attack Type')
plt.show()
```

```python
plt.subplots(figsize=(16,6))
sns.countplot(df['targtype1_txt'], data=df,order= df['targtype1_txt'].value_counts().index[:5])
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.xlabel("Target Type")
plt.title('Maximum Terrorism Target ')
plt.show()
```

```python
df_copy=df.copy()
df_copy["success"] = df["success"].replace({1:"Successful Operation", 0:"Failure Operation"})
plt.subplots(figsize=(16,6))
sns.countplot(df_copy['success'], data=df_copy,order= df_copy['success'].value_counts().index[:5])
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.xlabel("success Rate")
plt.title('Number of succes and Unsuccess attack ')
plt.show()
```

```python
Killedpeople = df[['iyear',"nkill"]].groupby(['iyear']).idxmax()
fig,axes = plt.subplots(figsize=(20,10))
c = ['red', 'yellow', 'black', 'blue', 'orange']
Killedpeople.plot(kind='bar',alpha=0.7,ax=axes,color=c)
plt.ylabel("No of people killed",fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.title("Killed Graph")
```
